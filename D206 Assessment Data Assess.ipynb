```python
import pandas as pd
import numpy as np
import missingno as msno
import seaborn as sns
```

```python
# import our csv file
file = pd.read_csv("churn_raw_data.csv")
```

```python
# check and print % nullity
file_nullity = file.isnull()
sum_missing = file_nullity.sum()
percent_missing = file_nullity.mean()*100
sum_missing = sum_missing[sum_missing != 0]
percent_missing = percent_missing[percent_missing != 0]
print("Number missing:\n",sum_missing,
     "\nPercent missing:\n",percent_missing)
```

```python
# a matrix to see where our missing values are
msno.matrix(file)
```

```python
# a heatmap to see any correlating missing data
msno.heatmap(file)
```

```python
# checking how we need to clean bandwidth_GB_per_year
# if the NA values are associated with accounts w/ no internet, set to 0
# if NA values are associated with accounts w/ internet, impute w/ avg
GB_compare = file[["InternetService", "Bandwidth_GB_Year"]]
msno.matrix(GB_compare)
```
